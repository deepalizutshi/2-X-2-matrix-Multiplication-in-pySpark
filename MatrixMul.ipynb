```
from pyspark.sql.types import *
from pyspark import SparkContext
from itertools import product
sc=SparkContext.getOrCreate()
import numpy as np

a=sc.textFile("dbfs:/FileStore/shared_uploads/deepali.zutshi@sjsu.edu/matrixA-7.txt").flatMap(lambda line: line.split(" ")).map(lambda x: int(x))
dataA=a.collect()

matrixA=[]
for k in (0,1):
  c=0
  for i in (0,1):
    for j in (0,1):
      matrixA.append(((i,k),(j,dataA[c])))
      c=c+1



b=sc.textFile("dbfs:/FileStore/shared_uploads/deepali.zutshi@sjsu.edu/matrixB-7.txt").flatMap(lambda line:line.split(" ")).map(lambda x: int(x))
dataB=b.collect()
matrixB=[]
for i in (0,1):
  c=0
  for j in (0,1):
    for k in (0,1):
      matrixB.append(((i,k),(j,dataB[c])))
      c=c+1


data=sc.parallelize(matrixA+matrixB).groupByKey().mapValues(sorted)    
data2 = data.map(lambda x: x[1][0][1]*x[1][1][1]+x[1][2][1]*x[1][3][1]).collect()
print([data2[1],data2[0]],'\n',[data2[2],data2[3]])
```
